## The algorithm

```python
def SCINT_test(X,Y,Z,M,nperm=500):
    Del_X = np.array([X[i]-X[j] for (i,j) in M])
    Del_Y = np.array([Y[i]-Y[j] for (i,j) in M])
    
    T   = Del_X @ Del_Y
    T_S = np.einsum('ij, j, j -> i', 
                    2*(np.random.rand(nperm, len(M)) < .5)-1, 
                    Del_X, 
                    Del_Y)
    
    return np.mean(T_S > T)

# def SCINT_test(X,Y,Z,M,nperm=500):
#     Del_X = [X[i]-X[j] for (i,j) in M]
#     Del_Y = [Y[i]-Y[j] for (i,j) in M]
    
#     T=np.dot(Del_X,Del_Y)
            
#     T_S = []
#     for t in range(nperm):
#         S = (np.random.rand(len(M))<0.5)*2-1        
#         T_S.append(np.dot(np.multiply(S,Del_X),Del_Y))
#     return np.mean([a>T for a in T_S])
```

## Simulation Settings

```python
def mu(Y,Z,beta):
    return (Z + beta * Y)
def simulation(n,beta,setting):
    if setting == 1:
        Z = np.random.normal(0,1,n)
        Y = np.random.normal(0,1,n)
        
    elif setting == 2:
        Z = np.random.normal(0,1,n)
        Y = Z + np.random.normal(0,1,n)
        
    elif setting == 3:
        Z = np.random.normal(0,1,n)
        Y = -Z + np.random.normal(0,1,n)
        
    X = mu(Y,Z,beta) + np.random.normal(0,1,n)
    return X,Y,Z
```

```python
beta=1/np.sqrt(2000)
X,Y,Z = simulation(2000,beta,2)
print(np.corrcoef(Y,Z))
plt.figure(figsize=(10,5))
plt.plot(Z, Y, '.', c=lblue)
plt.xticks([])
plt.yticks([])

plt.show()
```

# Question: Effect of dependence of $P_{Y,Z}$, and $(\beta_n)_n$ on power

```python
nrep=500;alpha=0.1

def power_analysis(setting):
    nseq=np.arange(500,30501,2000)
    power_nbhd_0=[];power_sb_0=[];power_cb_0=[]
    power_nbhd_1=[];power_sb_1=[];power_cb_1=[]
    power_nbhd_2=[];power_sb_2=[];power_cb_2=[]
    
    for n in tqdm(nseq):
        
        #below detection threshold
        beta=2/n
        pvalue_nbhd=[];pvalue_sb=[];pvalue_cb=[]
        for l in range(nrep):
            X,Y,Z = simulation(n,beta,setting)
            eta = 1/(n**0.75)
            M_nbhd = immediate_neighbor_matching(Y, Z)
            M_sb = same_bin_matching(Y, Z, eta)
            M_cb = cross_bin_matching(Y, Z, eta)
            pvalue_nbhd.append(SCINT_test(X,Y,Z,M_nbhd))
            pvalue_sb.append(SCINT_test(X,Y,Z,M_sb))
            pvalue_cb.append(SCINT_test(X,Y,Z,M_cb))
        power_nbhd_0.append(np.mean([p<alpha for p in pvalue_nbhd]))
        power_sb_0.append(np.mean([p<alpha for p in pvalue_sb]))
        power_cb_0.append(np.mean([p<alpha for p in pvalue_cb]))
    
        
        #at detection threshold
        beta=2/np.sqrt(n)
        pvalue_nbhd=[];pvalue_sb=[];pvalue_cb=[]
        for l in range(nrep):
            X,Y,Z = simulation(n,beta,setting)
            eta = 1/(n**0.75)
            M_nbhd = immediate_neighbor_matching(Y, Z)
            M_sb = same_bin_matching(Y, Z, eta)
            M_cb = cross_bin_matching(Y, Z, eta)
            pvalue_nbhd.append(SCINT_test(X,Y,Z,M_nbhd))
            pvalue_sb.append(SCINT_test(X,Y,Z,M_sb))
            pvalue_cb.append(SCINT_test(X,Y,Z,M_cb))
        power_nbhd_1.append(np.mean([p<alpha for p in pvalue_nbhd]))
        power_sb_1.append(np.mean([p<alpha for p in pvalue_sb]))
        power_cb_1.append(np.mean([p<alpha for p in pvalue_cb]))
    
        #above detection threshold
        beta=1/(n**0.42)
        pvalue_nbhd=[];pvalue_sb=[];pvalue_cb=[]
        for l in range(nrep):
            X,Y,Z = simulation(n,beta,setting)
            eta = 1/(n**0.75)
            M_nbhd = immediate_neighbor_matching(Y, Z)
            M_sb = same_bin_matching(Y, Z, eta)
            M_cb = cross_bin_matching(Y, Z, eta)
            pvalue_nbhd.append(SCINT_test(X,Y,Z,M_nbhd))
            pvalue_sb.append(SCINT_test(X,Y,Z,M_sb))
            pvalue_cb.append(SCINT_test(X,Y,Z,M_cb))
        power_nbhd_2.append(np.mean([p<alpha for p in pvalue_nbhd]))
        power_sb_2.append(np.mean([p<alpha for p in pvalue_sb]))
        power_cb_2.append(np.mean([p<alpha for p in pvalue_cb]))
        
    return power_nbhd_0,power_nbhd_1,power_nbhd_2,power_sb_0,power_sb_1,power_sb_2,power_cb_0,power_cb_1,power_cb_2

power_nbhd_10,power_nbhd_11,power_nbhd_12,power_sb_10,power_sb_11,power_sb_12,power_cb_10,power_cb_11,power_cb_12 = power_analysis(1)

power_nbhd_20,power_nbhd_21,power_nbhd_22,power_sb_20,power_sb_21,power_sb_22,power_cb_20,power_cb_21,power_cb_22 = power_analysis(2)

power_nbhd_30,power_nbhd_31,power_nbhd_32,power_sb_30,power_sb_31,power_sb_32,power_cb_30,power_cb_31,power_cb_32 = power_analysis(3)
```

# Question: Effect of different $\alpha$, and monotonicity on type-I error

```python
nrep=500

def effect_on_validity(alpha):
    nseq=np.arange(500,20501,2000)

    power_nbhd_0=[];power_sb_0=[];power_cb_0=[]
    power_nbhd_1=[];power_sb_1=[];power_cb_1=[]
    power_nbhd_2=[];power_sb_2=[];power_cb_2=[]
    
    for n in tqdm(nseq):
        pvalue_nbhd_0=[];pvalue_sb_0=[];pvalue_cb_0=[]
        pvalue_nbhd_1=[];pvalue_sb_1=[];pvalue_cb_1=[]
        pvalue_nbhd_2=[];pvalue_sb_2=[];pvalue_cb_2=[]
        
        for l in range(nrep):
            Z = np.random.normal(0,1,n)
            Y = np.random.normal(0,1,n)
            X1 = Z + np.random.normal(0,1,n)
            X2 = norm.cdf(Z)+ np.random.normal(0,1,n)
            X3 = 10*Z + np.random.normal(0,1,n)
            eta = 1/(n**0.75)
            M_nbhd = immediate_neighbor_matching(Y, Z)
            M_sb = same_bin_matching(Y, Z, eta)
            M_cb = cross_bin_matching(Y, Z, eta)
            
            pvalue_nbhd_0.append(SCINT_test(X1,Y,Z,M_nbhd))
            pvalue_sb_0.append(SCINT_test(X1,Y,Z,M_sb))
            pvalue_cb_0.append(SCINT_test(X1,Y,Z,M_cb))
            
            pvalue_nbhd_1.append(SCINT_test(X2,Y,Z,M_nbhd))
            pvalue_sb_1.append(SCINT_test(X2,Y,Z,M_sb))
            pvalue_cb_1.append(SCINT_test(X2,Y,Z,M_cb))
    
            pvalue_nbhd_2.append(SCINT_test(X3,Y,Z,M_nbhd))
            pvalue_sb_2.append(SCINT_test(X3,Y,Z,M_sb))
            pvalue_cb_2.append(SCINT_test(X3,Y,Z,M_cb))
    
        
        power_nbhd_0.append(np.mean([p<alpha for p in pvalue_nbhd_0]))
        power_sb_0.append(np.mean([p<alpha for p in pvalue_sb_0]))
        power_cb_0.append(np.mean([p<alpha for p in pvalue_cb_0]))
    
        power_nbhd_1.append(np.mean([p<alpha for p in pvalue_nbhd_1]))
        power_sb_1.append(np.mean([p<alpha for p in pvalue_sb_1]))
        power_cb_1.append(np.mean([p<alpha for p in pvalue_cb_1]))
    
        power_nbhd_2.append(np.mean([p<alpha for p in pvalue_nbhd_2]))
        power_sb_2.append(np.mean([p<alpha for p in pvalue_sb_2]))
        power_cb_2.append(np.mean([p<alpha for p in pvalue_cb_2]))

    return power_nbhd_0,power_nbhd_1,power_nbhd_2,power_sb_0,power_sb_1,power_sb_2,power_cb_0,power_cb_1,power_cb_2

power_nbhd_10,power_nbhd_11,power_nbhd_12,power_sb_10,power_sb_11,power_sb_12,power_cb_10,power_cb_11,power_cb_12 = effect_on_validity(0.01)

power_nbhd_20,power_nbhd_21,power_nbhd_22,power_sb_20,power_sb_21,power_sb_22,power_cb_20,power_cb_21,power_cb_22 = effect_on_validity(0.05)

power_nbhd_30,power_nbhd_31,power_nbhd_32,power_sb_30,power_sb_31,power_sb_32,power_cb_30,power_cb_31,power_cb_32 = effect_on_validity(0.1)

```
